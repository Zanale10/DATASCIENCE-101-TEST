````
import requests
import pandas as pd

API_KEY = '65066d8f6ece43cebb690836251206'
BASE_URL = 'https://api.weatherapi.com/v1/current.json'

# List of cities
cities = ['Nairobi', 'Kampala', 'Mombasa', 'Addis Ababa', 'Dar es Salaam',
          'Lusaka', 'Kigali', 'Johannesburg', 'Cairo', 'Lagos']

# Store weather data for all cities
weather_data = []

for city in cities:
    params = {
        'key': API_KEY,
        'q': city
    }

    response = requests.get(BASE_URL, params=params)

    if response.status_code == 200:
        data = response.json()
        current = data['current']

        weather_data.append({
            'city': city,
            'temperature (C)': current['temp_c'],
            'humidity (%)': current['humidity'],
            'pressure (mb)': current['pressure_mb'],
            'wind_speed (kph)': current['wind_kph']
        })
    else:
        print(f"Failed to get data for {city}: {response.status_code}")

# Convert to DataFrame and print
df = pd.DataFrame(weather_data)
print(df)
````

```python
import streamlit as st
import pandas as pd
from datetime import datetime
import os

# ğŸ” Path to your OneDrive-synced Excel file
excel_file_path = r"C:\Users\YOUR_NAME\OneDrive\ProductionData.xlsx"

st.title("ğŸ“‹ Production Data Entry")

# Form Inputs
with st.form("data_entry_form"):
    date = st.date_input("Date")
    shift = st.selectbox("Shift", ["Morning", "Afternoon", "Night"])
    machine = st.text_input("Machine")
    supervisor = st.text_input("Supervisor")
    operator = st.text_input("Operator")
    output = st.number_input("Output", min_value=0, step=1)
    scrap = st.number_input("Scrap", min_value=0, step=1)
    reason for scrap = st.text_input("Reason for Scrap")
    Machine downtime= st.number_input("Machine downtime", min_value=0, step=1)
    submitted = st.form_submit_button("Submit")

    if submitted:
        new_entry = pd.DataFrame([{
            "Date": date.strftime('%Y-%m-%d'),
            "Shift": shift,
            "Machine": machine,
            "Supervisor": supervisor,
            "Operator": operator,
            "Output": output,
            "Scrap": scrap,
            "Reason for Scrap": reason for scrap,
            "Machine downtime": Machine downtime,
            "Timestamp": datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        }])

        # Load or create Excel
        if os.path.exists(excel_file_path):
            existing_data = pd.read_excel(excel_file_path)
            updated_data = pd.concat([existing_data, new_entry], ignore_index=True)
        else:
            updated_data = new_entry

        updated_data.to_excel(excel_file_path, index=False)
        st.success("âœ… Data submitted and saved to OneDrive Excel!")
```
